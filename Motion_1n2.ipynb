```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
```

```python
ttwo, xtwo, vtwo, atwo, ptwo = np.loadtxt("Data h=2.0.csv", skiprows=3, unpack=True, delimiter=",")
tfiv, xfiv, vfiv, afiv, pfiv = np.loadtxt("Data h=5.4.csv", skiprows=3, unpack=True, delimiter=",")
```

```python
figure, axis = plt.subplots(3, 1)

axis[0].plot(ttwo, xtwo)
#axis[0].ylabel("Distance(cm)")
axis[1].plot(ttwo, vtwo)
axis[2].plot(ttwo, atwo)
axis[0].set_xlabel('Time(s)')
axis[1].set_xlabel('Time(s)')
axis[2].set_xlabel('Time(s)')
axis[0].set_ylabel('Position(m)')
axis[1].set_ylabel('Velocity(m/s)')
axis[2].set_ylabel('Acceleration(m/s^2)')
```

```python
figure, axis = plt.subplots(3, 1)

axis[0].plot(tfiv, xfiv)
#axis[0].ylabel("Distance(cm)")
axis[1].plot(tfiv, vfiv)
axis[2].plot(tfiv, afiv)
axis[0].set_xlabel('Time(s)')
axis[1].set_xlabel('Time(s)')
axis[2].set_xlabel('Time(s)')
axis[0].set_ylabel('Position(m)')
axis[1].set_ylabel('Velocity(m/s)')
axis[2].set_ylabel('Acceleration(m/s^2)')
```

```python
def GetVelocity(x_1,x_2):
  Velocity = (x_2 - x_1)/(1/25)
  return round(Velocity,3)


x2_v = xtwo[2:]

VelocityList = []
for (x_1,x_2) in zip(xtwo,x2_v):
  VelocityList.append(GetVelocity(x_1,x_2))

print(VelocityList)
print(vtwo)
```

```python
def GetAcceleration(x_1,x_2,x_3):
  Acceleration = (x_3 - 2*x_2 + x_1)/(1/50)**2
  return Acceleration

a_x_1 = xtwo
a_x_2 = xtwo[1:]
a_x_3 = xtwo[2:]

AccelerationList = []
for (x_1,x_2,x_3) in zip(a_x_1,a_x_2,a_x_3):

  AccelerationList.append(GetAcceleration(x_1,x_2,x_3))

print(AccelerationList)
print(atwo)
```
